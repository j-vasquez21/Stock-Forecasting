```python
symbols = ['NASDX','QQQ','VOO','SPY','VTWO','IVV','SWPPX','DIA','VFIAX','VTSAX','FXAIX','SWTSX','VUG']
```

```python
resampled = []
for x,y in zip(funds_list,symbols):
    x = x.reset_index()
    x['date'] = pd.to_datetime(x['date'])
    x = x.set_index(['symbol','date'])
    x = x.loc[y].resample('D').mean().bfill()
    x['symbol'] = pd.Series(y)
    x['symbol'] =  x['symbol'].fillna(y)
    x = x.reset_index().set_index(['symbol','date'])
    resampled.append(x)

for x in resampled:
    funds = pd.concat([funds, x])
```

```python
def check_stationarity(ts):
    #Determing rolling statistics
    roll_mean = ts.rolling(12).mean()
    roll_std = ts.rolling(12).std()
    #Plot rolling statistics:
    plt.plot(ts, color='blue',label='Original')
    plt.plot(roll_mean, color='red', label='Rolling Mean')
    plt.plot(roll_std, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    
    print("Results of dickey fuller test")
    adft = adfuller(ts,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
```
